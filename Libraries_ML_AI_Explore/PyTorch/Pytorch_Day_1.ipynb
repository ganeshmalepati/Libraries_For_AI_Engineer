```python
x = torch.tensor(5.0, requires_grad=True)
y = x ** 2 + 2*x + 3*55
z = y * 5
print(y)
print(z)
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x * 3
z = y + 4
print(x)
print(y)
print(z)
print(z.grad_fn)
```

```python
x = torch.tensor(5.0, requires_grad=True)

y = x ** 3 + 3*x + 25

y.backward()

print(y.grad)
```

```python
x = torch.tensor(2.0, requires_grad=True)

y = x * 2
z = y * 5

z.backward()

print(x.grad)
```

```python
x = torch.tensor(2.0, requires_grad=True)

with torch.no_grad():
    y = x * 2
print(y.requires_grad)
```

```python
w=torch.tensor(2.0, requires_grad=True)

x= torch.tensor(4.0)

y = w * x

loss = (y - 10) ** 2

loss.backward()

with torch.no_grad():
    w -= 0.01 * w.grad
    w.grad.zero_()

print(w)
print(w.requires_grad)
print(w.grad)
print(w.grad_fn)
```
