```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np 
```

```python
sent_length = 8
embedding_doc = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
embedding_doc
```

```python
dim = 10
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam', 'mse')
```

```python
import numpy as np 
import pandas as pd 
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
```

```python
word_idx=imdb.get_word_index()
reverse_word_idx={value:key for key, value in word_idx.items()}
```

```python
model = load_model('simple_rnn_imdb.h5')
model.summary()
```

```python
model.get_weights()
```

```python
def decode_review(encoded_review):
    return ' '.join(reverse_word_idx.get(i - 3, '?') for i in encoded_review)

def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_idx.get(word, 2) + 3 for word in words]
    padded_seq = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_seq

```

```python
def prediction_sent(review):
    preprocessed_text = preprocess_text(review)
    prediction=model.predict(preprocessed_text)
    sentement = 'Positive' if prediction[0][0]  > 0.5 else 'Negative'
    return sentement, prediction[0][0]
```

```python
example_review = "This movie was fantastic! The acting was great and the plot was thrilling."

sentement,score=prediction_sent(example_review)
print(f"Review: {example_review}")
print(f"Sentement: {sentement}")
print(f"Score: {score}")
```
