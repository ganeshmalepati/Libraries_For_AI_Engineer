```python
import numpy as np
import pandas as pd
# import matplotlib as mp
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
)
```

```python
def load_and_explore_dataset():
    print("\n" +"="*60)
    print("STEP 0: Data Loading")
    print("\n" +"="*60)

    data = load_breast_cancer()
    X = data.data
    y = data.target

    print(f"\n Shape of the data: {X.shape}")
    print(f"\n Number of labels: {len(np.unique(y))}")
    print(f"\n Class names: {data.target_names}")
    print(f"\n Malignant (0): {np.sum(y == 0)} samples")
    print(f"\n Bengin (1): {np.sum(y == 1)} samples")
    print("\n Feature names: ")
    for i, name in enumerate(data.feature_names):
        print(f" [{i:2d}] {name}")
    df = pd.DataFrame(X, columns=data.feature_names)
    df["label"] = y
    print(f"\n First 3 rows: \n {df.head(3).to_string()} \n")

    return X, y, data.feature_names
```

```python
load_and_explore_dataset()
```
